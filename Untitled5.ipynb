```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file = pd.read_csv('hw_stat_2018.csv', delimiter=';')
```

```python
file.head()
```

```python
filtered_file = file.loc[(file['WallDurationSeconds'] > 5000) & (file['WallDurationSeconds'] < 60000)]
wall_d = filtered_file['WallDurationSeconds'].tolist()
print(len(wall_d))
```

```python
ax = sns.distplot(wall_d, hist=False, kde=True)
plt.savefig("plot.jpg")
```
